```python
plt.figure(figsize=(14,8))
ax = app_train['CNT_CHILDREN'].value_counts().plot(kind='bar',title='CNT_CHILDREN')
ax.set_xlabel('No. of Children')
ax.set_ylabel('Frequency Count')
print('There are outliers or really rare cases in children count')
```

```python
plt.figure(figsize=(14,8))
ax = app_train['NAME_CONTRACT_TYPE'].value_counts().plot(kind='bar',title='NAME_CONTRACT_TYPE')
ax.set_xlabel('Type of Contract')
ax.set_ylabel('Frequency Count')
print("The contract type are mostly cash loans")
```

```python
plt.figure(figsize=(14,8))
ax = app_train['CODE_GENDER'].value_counts().plot(kind='bar')
ax.set_xlabel('Gender')
ax.set_ylabel('Frequency Count')
print("The number of male is double the number of female")
```

```python
plt.figure(figsize=(14,8))
ax = app_train['AMT_GOODS_PRICE'].hist(bins=200)
ax.set_xlabel('Goods Price')
ax.set_ylabel('Frequency Count')
ax.set_title('AMT_GOODS_PRICE')
print("The goods price are mostly around 500000")
```

```python
plt.figure(figsize=(14,8))
dat = app_train['AMT_INCOME_TOTAL']
plt.plot(np.arange(len(dat)),dat)
plt.xlabel('Data Index')
plt.ylabel('Total Income')
plt.title('AMT_INCOME_TOTAL')
print('There is a guy with crazy income or an invalid outlier data')
```
