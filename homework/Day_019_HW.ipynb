```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件

# 忽略警告訊息
%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')

# 設定 data_path
dir_data = './data/'
```

```python
# 讀取檔案
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
app_train = app_train[~app_train['HOUSETYPE_MODE'].isnull()]
# app_train['YEARS_BIRTH'] = abs(app_train['DAYS_BIRTH'])/365
app_train = app_train[app_train['AMT_INCOME_TOTAL'] != 117000000.0]
app_train['AMT_INCOME_TOTAL'] = np.log10(app_train['AMT_INCOME_TOTAL'])
```

```python
# 使用不同的 HOUSETYPE_MODE 類別繪製圖形, 並使用 subplot 排版
"""
Your Code Here
"""
unique_house_type = app_train['HOUSETYPE_MODE'].unique()

nrows = len(unique_house_type)
# ncols = nrows // 2
ncols = 2

plt.figure(figsize=(15,20))
for i in range(len(unique_house_type)):
    plt.subplot(nrows, ncols, i+1)
    """
    Your Code Here
    """
    app_train.loc[app_train['HOUSETYPE_MODE'] == unique_house_type[i],'AMT_CREDIT'].hist(bins=25,edgecolor='k')
    plt.xlabel('AMT_CREDIT')
    plt.ylabel('Number of house_type')
    plt.title(str(unique_house_type[i]))
    
plt.show()    
```

```python
# 使用不同的 HOUSETYPE_MODE 類別繪製圖形, 並使用 subplot 排版

plt.figure(figsize=(15,20))
for i in range(len(unique_house_type)):
    plt.subplot(nrows, ncols, i+1)
    """
    Your Code Here
    """
    app_train.loc[app_train['HOUSETYPE_MODE'] == unique_house_type[i],'AMT_INCOME_TOTAL'].hist(bins=25,edgecolor='k')
    plt.xlabel('Normalized Total Income')
    plt.ylabel('Number of house_type')
    plt.title(str(unique_house_type[i]))
    
plt.show()    
```
